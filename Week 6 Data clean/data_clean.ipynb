```python
bilibili_data = {'video_url': [], 'city': [],  'video_name': [], 
                 'interaction_count': [], 'comment_count': [],
                 'comment_min_len': [], 'comment_max_len': [], 
                 'comment_mean_len': [], 'comment_std_len': []}
regobj = re.compile('^\[.*\]$')
regobj2 = re.compile('^def.*')
for prov_ind, prov_name in enumerate(province_names):
    for city_ind, city_name in enumerate(city_names[prov_ind]):
        for url, video in city_data_dict[city_name].items():
            bilibili_data['city'].append(city_name)
            bilibili_data['video_url'].append(url)
            bilibili_data['video_name'].append(video[1])
            bilibili_data['interaction_count'].append(video[2]['interactionStatistic']['userInteractionCount'])
            comments = video[3]
            # Filter the comments having format control content
            comments = [s for s in comments if regobj.match(s) is None and regobj2.match(s) is None]
            comment_lens = [len(comment) for comment in comments]
            if len(comments) != 0:
                bilibili_data['comment_count'].append(len(comments))
                bilibili_data['comment_min_len'].append(np.min(comment_lens))
                bilibili_data['comment_mean_len'].append(np.mean(comment_lens))
                bilibili_data['comment_std_len'].append(np.std(comment_lens))
                bilibili_data['comment_max_len'].append(np.max(comment_lens))
            else:
                bilibili_data['comment_count'].append(0)
                bilibili_data['comment_min_len'].append(np.nan)
                bilibili_data['comment_mean_len'].append(np.nan)
                bilibili_data['comment_std_len'].append(np.nan)
                bilibili_data['comment_max_len'].append(np.nan)
bilibili_df = pd.DataFrame(bilibili_data)
```

```python
fontfamily = 'SimSun'
city_num = 5
bins_interaction = np.concatenate([np.arange(0, 1.0, 0.2) * 4202, [11583500]], axis=0).astype(np.int)
bins_comment = np.concatenate([np.arange(0, 1.0, 0.2) * 14, [8000]], axis=0).astype(np.int)
fig = plt.figure(figsize=(10, 2*city_num))
axs = fig.subplots(city_num, 2, sharex='col', squeeze=False)
row = 0
col = 0
def custom_hist(df, bins, ax):
    counts, _ = np.histogram(df, bins=bins, range=(0, bins[-1]))
    ax.bar(bins[:-1], counts, width=0.8*(bins[1]-bins[0]), align='edge', tick_label=bins[:-1])
    ax.set_xlim(0, bins[-2]+1.0*(bins[1]-bins[0]))
for prov_ind, prov_name in enumerate(province_names):
    for city_ind, city_name in enumerate(city_names[prov_ind]):
        if row >= city_num:
            continue
        custom_hist(bilibili_df[bilibili_df['city'] == city_name].iloc[:, 3], bins_interaction, axs[row, col])
        custom_hist(bilibili_df[bilibili_df['city'] == city_name].iloc[:, 4], bins_comment, axs[row, col+1])
        axs[row, col].set_ylabel(city_name, fontfamily=fontfamily, fontsize=15)
        row += 1
axs[-1, 0].set_xlabel('Interaction Count Histogram', fontfamily='serif', fontsize=15)
axs[-1, 1].set_xlabel('Comment Count Histogram', fontfamily='serif', fontsize=15)
pass
```

```python
comment_data = {}
regobj = re.compile('^\[.*\]$')
regobj2 = re.compile('^def.*')
for prov_ind, prov_name in enumerate(province_names):
    for city_ind, city_name in enumerate(city_names[prov_ind]):
        comment_data[city_name] = {'interaction_count': [], 'comments': [], 'comment_count': [], 'comment_lens': []}
        for url, video in city_data_dict[city_name].items():
            comment_data[city_name]['interaction_count'].append(video[2]['interactionStatistic']['userInteractionCount'])
            comments = video[3]
            # Filter the comments having format control content
            comments = [s for s in comments if regobj.match(s) is None and regobj2.match(s) is None]
            comment_lens = [len(comment) for comment in comments]
            comment_data[city_name]['comments'].append(comments)
            comment_data[city_name]['comment_count'].append(len(comments))
            comment_data[city_name]['comment_lens'].append(comment_lens)
            if len(comment_lens) > 0 and np.max(comment_lens) > 100:
                print(url, comments, video[0])
                print('\n\n')
```

```python
interested_cities = ['海口', '玉林', '牡丹江', '阳泉', '长治', '台北', '高雄']
fig = plt.figure(figsize=(10, 2*len(interested_cities)), linewidth=0.0, edgecolor='c')
axs = fig.subplots(len(interested_cities), 1, sharex='col', squeeze=False)
fig.subplots_adjust(top=1.0, right=1.0, bottom=0.04, left=0.08)
row = 0
sns.set_style('white')
total_comment_lens = []
for city_name in interested_cities:
    comment_lens = np.concatenate(comment_data[city_name]['comment_lens'], axis=0)
    total_comment_lens.extend(comment_lens)
    print('Mean:{} Std:{}'.format(np.mean(comment_lens), np.std(comment_lens)))
    sns.distplot(comment_lens, ax=axs[row, 0], rug=True)
    axs[row, 0].set_ylabel(city_name, fontfamily='SimSun', fontsize=15)
    row += 1
axs[-1, 0].set_xlabel('Comment Lens Bins', fontsize=15)
plt.savefig('histogram of comment length for popular cities.png', dpi=400)
print('Total mean: {} Total std: {}'.format(np.mean(total_comment_lens), np.std(total_comment_lens)))
```
